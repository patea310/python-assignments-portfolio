```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Density(mean, w_sqrd, x): # Defining for f_x(x), will give the probablity of the existence of the inputted x
    
    Output = ((1/np.sqrt(2*np.pi*w_sqrd))*np.e**(-1*((x-mean)**2)/(2*w_sqrd))) # Computation for output of f_x(x), 
    
    # Iput of 3 variables
    
    return Output # Outputting
    
def Density_Plot(mean, w_sqrd): # Defining the function to plot x in [c,d]
    
    c = 1
    d = 10
    n = 100
    x = np.linspace(c,d,n+1) # Using linspace to create a list with n number of x
   
    Output = [] # List to hold y
   
    for i in x: # Iterate through the list of x's to append to y list
        Output.append((1/np.sqrt(2*np.pi*w_sqrd))*np.e**(-1*((i-mean)**2)/(2*w_sqrd)))
    return x, Output
    
x,Density_Values=Density_Plot(5,9.7) 

# Likelihood of a number being between 1-10 when the mean is 5, and the standard deviation^2 is 9.7

# Scheming and Plotting :))
plt.scatter(x, Density_Values, label ="Normal_Density") 
plt.xlabel("x")
plt.ylabel("y")
plt.title("Normal Distribution")
plt.legend()
plt.grid()
plt.show()
```
